````python
_PATH_TO_DATA_SOURCE = '../data/source/'
_PATH_TO_DATA_RAW = '../data/raw/'
list_files = os.listdir(_PATH_TO_DATA_SOURCE)
filename = _PATH_TO_DATA_SOURCE+list_files[0]
print(filename)
data_source = pd.read_csv(filename)
data_source
````

````python
data_source.columns
````

````python
NB_PARTITIONS = 10
size_partitions = int(np.ceil(len(data_source)/NB_PARTITIONS))
for i in range(NB_PARTITIONS):
    print(i)
    data_i = data_source.iloc[i*size_partitions:i*size_partitions+size_partitions+5].drop(columns=['Sleep Disorder'])
    display(data_i)
    filename_i = _PATH_TO_DATA_RAW+re.sub('.csv','_'+str(i)+'.csv',list_files[0])
    print(filename_i)
    data_i.to_csv(filename_i,index=False)
    
    
data_i = data_source[['Person ID','Sleep Disorder']]
display(data_i)
filename_i = _PATH_TO_DATA_RAW+'Target_Sleep_Disorder.csv'
print(filename_i)
data_i.to_csv(filename_i,index=False)
````
